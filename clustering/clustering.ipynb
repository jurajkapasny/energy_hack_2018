```python
# Juraj
data_path = "/Users/jurajkapasny/Data/energy_hack/"
om_data = pd.read_csv(data_path+"spotreba_prepared.csv",sep = ";")
```

```python
om_data.head(10)
```

```python
om_data.columns = ["date","interval","spotreba","om"]
```

```python
om_data['interval'] = om_data['date'] + ' ' + om_data['interval']
```

```python
om_data.head()
```

```python
om_data.interval = pd.to_datetime(om_data.interval)
om_data['hour'] = om_data.interval.dt.hour
```

```python
df = om_data.groupby([om_data.om, 'hour'])['spotreba'].sum().unstack()
```

```python
df.tail(5)
```

```python
df_norm = (df - df.mean()) / (df.max() - df.min())
```

```python
df_norm.head()
```

```python
dataset = df_norm.values
```

```python
km = KMeans(n_clusters = 4)
```

```python
km.fit(dataset)
```

```python
x = km.fit_predict(dataset)
```

```python
df['cluster'] = x
```

```python
df.groupby("cluster").size()
```

```python
df.groupby("cluster").mean().transpose()
```

```python
df[["cluster"]].to_csv(data_path + "clusters.csv", sep = ";", index = True)
```
