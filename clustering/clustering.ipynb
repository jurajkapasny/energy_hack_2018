```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
om_data = pd.read_csv("/Volumes/doc/projects/energyhack2/data_prepared/spotreba_prepared.csv", sep=";")
```

```python
om_data.head(10)
```

```python
om_data['interval'] = om_data['date'] + ' ' + om_data['interval']
```

```python
om_data
```

```python
om_data.interval = pd.to_datetime(om_data.interval)
om_data['hour'] = om_data.interval.dt.hour
```

```python
df = om_data.groupby([om_data.om, 'hour'])['spotreba'].sum().unstack()
```

```python
df.tail(5)
```

```python
df_norm = (df - df.mean()) / (df.max() - df.min())
```

```python
df_norm
```

```python
dataset = df_norm.values
```

```python
km = KMeans(n_clusters = 3)
```

```python
km.fit(dataset)
```

```python
x = km.fit_predict(dataset)
```

```python
df['cluster'] = x
```
