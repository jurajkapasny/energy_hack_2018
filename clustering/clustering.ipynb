```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

Import data

```python
omd = pd.read_csv("/Volumes/doc/projects/energyhack2/data_prepared/spotreba_prepared.csv", sep=";")
```

```python
omd.head(10)
```

```python
short = omd[:100000]
```

```python
short
```

```python
short['interval'] = short['date'] + ' ' + short['interval']
```

```python
short
```

```python
short.interval = pd.to_datetime(short.interval)
short['hour'] = short.interval.dt.hour
```

```python
df = short.groupby([short.om, 'hour'])['spotreba'].sum().unstack()
```

```python
df
```

```python
df_norm = (df - df.mean()) / (df.max() - df.min())
```

```python
df_norm
```

```python
dataset = df_norm.values
```

```python
print(dataset)
```

```python
km = KMeans(n_clusters = 3)
```

```python
km.fit(dataset)
```

```python
x = km.fit_predict(dataset)
```

```python
df['cluster'] = x
```
